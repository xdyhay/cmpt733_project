```python
def collect_data(hashtags, start_date, end_date, name):
    for h in hashtags:
        tweet = twint.Config()
        tweet.Search = '#'+h
        tweet.Since = start_date
        tweet.Until = end_date
        tweet.Pandas = True
        tweet.Lang = 'en'
        twint.run.Search(tweet)
        df = twint.storage.panda.Tweets_df
        date_only = start_date.split('-')[1]+start_date.split('-')[2]
        twint.output.panda.Tweets_df.to_csv(name+'/'+date_only+'/'+h+'.csv', index=False)
```

```python
collect_data(conspiracy_hashtags, '2022-01-20', '2022-01-22', 'conspiracy_theories_data')
```
